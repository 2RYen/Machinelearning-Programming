```python
import pandas as pd  # (1) pandas 모듈 호출

# (2) 데이터 URL을 변수에 저장 (줄바꿈 오류 수정)
data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data'

# (3) 공백 구분자로 데이터 로드 (헤더 없음)
df_data = pd.read_csv(data_url, sep='\s+', header=None) #csv 형태의 파일로 불러오겠다. xls의 형식은 read_exceㅣ()
#sep- sperated
#\s+라고 하는 것은 정규 표현식의 의미
#특수문자 \를 사용하고,
# \n은 개행문자(새 줄 문자)
# \t는 탭문자(일정 공간만큼 옆을 공백 넣어줌)
# \s 공백문자(whitespace)
print(df_data.head())
```

```python
# (4) 열 이름 지정
df_data.columns = [
    'CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE',
    'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'
]

# (5) 데이터 상위 5개 출력
# 데이터의 형태와 구조를 알기 위해 처음 5개를 기본으로 출력해주는 역할.
# 데이터가 잘 불러와졌는지 눈으로 확인.
print(df_data.head())
```
