```python
from sklearn import datasets
iris = datasets.load_iris()

print(iris)
```

```python
print(iris.data[0]) #제일 첫번째에 있는 데이터가 불러와짐.
print(iris.target[0])
```

```python
from sklearn.model_selection import train_test_split

X = iris.data
y = iris.target

# (80:20)으로 분할한다.
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)

from sklearn import metrics
scores = metrics.accuracy_score(y_test, y_pred)

print(scores)
```

```python
classes = {0:'setosa',1:'versicolor',2:'virginica'}

# 전혀 보지 못한 새로운 데이터를 제시해보자.
x_new = [[3,4,5,2],
         [5,4,2,2]]

y_predict = knn.predict(x_new)

print(classes[y_predict[0]])
print(classes[y_predict[1]])
```

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
digits = datasets.load_digits()
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, digits.target, test_size=0.2, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(X_train, y_train)
```

```python
ConfusionMatrixDisplay.from_estimator(knn, X_test, y_test)
plt.show()
```

```python
y_pred = knn.predict(X_test)
print(f"{metrics.classification_report(y_test, y_pred)}\n")
```
