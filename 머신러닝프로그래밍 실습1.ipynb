```python
import numpy as np

# Ground truth & Predicted label
y_true = np.array([3.0, -0.5, 2.0, 7.0])
y_pred = np.array([2.5, 0.0, 2.0, 8.0])
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_true, y_pred)
print("MAE (sklearn): ", mae)
```

```python
mae_np = np.mean(np.abs(y_true - y_pred))
print("MAE (numpy): ", mae_np)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_true, y_pred)
print("MSE (sklearn): ", mse)
```

```python
mse_np = np.mean((y_true - y_pred)**2)
print("MSE (numpy): ", mse_np)
```

```python
#이건 원래 오류가 나오는게 맞음!! 괜히 고치려고 하지 말기!!
from sklearn.metrics import mean_squared_error
rmse = mean_squared_error(y_true, y_pred, squared=False)
print("RMSE (sklearn): ", rmse)
```

```python
rmse_np = np.sqrt(np.mean((y_true - y_pred) ** 2))
print("RMSE (numpy): ", rmse_np)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_true, y_pred)
print("R2 (sklearn): ", r2)
```

```python
ss_res = np.sum((y_true - y_pred) ** 2)
ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
r2_np = 1 - (ss_res / ss_tot)
print("R2 (numpy): ", r2_np)
```
